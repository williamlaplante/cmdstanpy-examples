```python
from cmdstanpy import CmdStanModel
import matplotlib.pyplot as plt
from scipy.stats import beta
import numpy as np
```

```python
path = './stanfiles/sir.stan' #Path to Stan model

raw_data = {"date":
         ['1978-01-22','1978-01-23','1978-01-24',
         '1978-01-25','1978-01-26','1978-01-27',
         '1978-01-28','1978-01-29','1978-01-30',
         '1978-01-31','1978-02-01','1978-02-02',
         '1978-02-03','1978-02-04'],
       "in_bed":
         [3,8,26,76,225,298,258,233,189,128,68,29,14,4],
       "convalescent":
         [0,0,0,0,9,17,105,162,176,166,150,85,47,20]   
}

data = {"n_days": 14, "N": 763, "y0": [762, 1, 0], "t0": 0,
        "ts": np.arange(1,15,1), "cases": raw_data["in_bed"]}
```

```python
model = CmdStanModel(stan_file = path)
```

```python
niter = 2000
chains = 4

fit = model.sample(data = data,
                   iter_sampling = niter,
                   chains = chains)
```
